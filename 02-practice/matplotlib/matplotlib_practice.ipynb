```python
import matplotlib.pyplot as plt
import numpy as np
```

## Matplotlib Basics

```python
x = np.arange(0, 10)
y = x * 2
print(x, y)
```

```python
plt.plot(x, y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xlim(0, 10)
plt.ylim(0, 20)
plt.title('String Title')
plt.savefig('myfirstplot.png')
plt.show()
```

## OOP Way

```python
a = np.linspace(0, 10, 11)
b = a ** 4
print(a, b)

x = np.arange(0, 10)
y = x * 2
print(x, y)
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)
axes1.set_xlim(0,)
axes1.set_ylim(0,)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of 4')

axes2 = fig.add_axes([0.1, 0.5, 0.4, 0.4])
axes2.plot(a, b)
axes2.set_xlim(6,8)
axes2.set_ylim(2000,4000)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed In')


plt.show()
```

```python
fig = plt.figure(dpi=100, figsize=(12,8))
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(x,y)
fig.savefig('newfig.png', bbox_inches="tight")
plt.show()
```

```python
fig, [ax1, ax2] = plt.subplots(nrows=2, ncols=1, figsize=(8,12), dpi=100)

ax1.plot(a, b)
ax1.set_title('A vs B')
ax1.set_xlabel('a')
ax1.set_ylabel('b')

ax2.plot(x, y)
ax2.set_title('X vs Y')
ax2.set_xlabel('x')
ax2.set_ylabel('y')


# plt.tight_layout()
fig.subplots_adjust(hspace=0.3)
fig.suptitle('Two Figures at Once', fontsize=15)
plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(x, x, label='X vs. X')
ax.plot(x, x**2, label='X vs. X^2')

ax.set_xlim(0,)
ax.set_ylim(0,)

ax.legend(loc='best')

plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(x, x, color='#650dbd', label='X vs. X')
ax.plot(x, x**2, color='red', linewidth=0.8, linestyle='--', marker='o', markersize='3', label='X vs. X^2')

plt.legend()
plt.show()
```
