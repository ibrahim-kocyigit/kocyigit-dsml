```python
import numpy as np
import matplotlib.pyplot as plt
```

### Linear Function

```python
# Generate x values: 100 evenly spaced points from -5 to 5.
x = np.linspace(-5, 5, 100)

# Calculate y values based on the linear equation y = 0.5x + 1.
y = 0.5 * x + 1

# Draw a horizontal line at y=0 (the x-axis).
plt.axhline(y=0, color='k')

# Draw a vertical line at x=0 (the y-axis).
plt.axvline(x=0, color='k')

# Add a grid to the plot for better readability.
plt.grid()

# Plot the linear function using the generated x and y values.
plt.plot(x, y)

# Display the plot.
plt.show()
```
