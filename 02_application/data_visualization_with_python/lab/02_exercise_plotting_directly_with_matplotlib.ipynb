```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df_can = pd.read_pickle('../data/interim/immigration_to_canada_01.pkl')
years = list(range(1980, 2014))
```

#### Task 1: Created a line plot to visualize the immigrants (to Canada) trend during 1980 to 2013

```python
immigration_per_year = df_can[years].transpose().sum(axis=1).to_frame().reset_index()
immigration_per_year.columns = ["year", "total"]
```

```python
fig, ax = plt.subplots()

ax.plot(immigration_per_year["year"], immigration_per_year["total"])
ax.set_title("Immigration to Canada from 1980 to 2013")
ax.set_xlabel("Year")
ax.set_ylabel("Number of Immigrants")
ax.set_ylim(0, 300000)
ax.grid(True)
ax.legend(["Immigrants"])
plt.show()
```

#### Task 2: Plot a line graph of immigration from Haiti

In 2010, Haiti suffered a catastrophic magnitude 7.0 earthquake. The quake caused widespread devastation and loss of life and aout three million people were affected by this natural disaster. As part of Canada's humanitarian effort, the Government of Canada stepped up its effort in accepting refugees from Haiti. We can quickly visualize this effort using a Line plot.

```python
df_haiti = df_can.set_index('country').loc[['Haiti']][years].transpose().reset_index()
df_haiti.columns = ['year', 'total']
```

```python
fig, ax = plt.subplots()

ax.plot(df_haiti['year'], df_haiti['total'], marker='o', markersize=3)
ax.set_title("Immigration from Haiti to Canada (1980 - 2013)")
ax.set_xlabel("Year")
ax.set_ylabel("Number of Immigrants")
ax.set_ylim(0, 7000)
ax.grid(True)
ax.annotate("2010 Earthquake", xy=(2000, 6000))
plt.show()
```

#### Task 3: Create a Scatter plot to visualize the immigrants (to Canada) trend during 1980 to 2013.

```python
fig, ax = plt.subplots()

ax.scatter(immigration_per_year['year'], immigration_per_year['total'], alpha=0.7)
ax.set_title("Immigration to Canada (1980 - 2013)")
ax.set_xlabel("Year")
ax.set_ylabel("Number of Immigrants")
plt.show()
```

#### Task 4: Create a bar plot to visualize the top 5 countries that contributed the most immigrants to Canada from 1980 to 2013.

```python
top_5 = df_can.sort_values(by='total', ascending=False).head(5)[['country', 'total']]
top_5
```

```python
fig, ax = plt.subplots()

labels = list(top_5['country'])
labels[2] = 'UK'

ax.bar(x=labels, height=top_5['total'])
ax.set_title("Immigration Trend of Top 5 Countries")
ax.set_ylabel('Number of Immigrants')
ax.set_xlabel('Years')

plt.show()
```
